### Challenge:

Automating a report with all the company's merchandise purchases, sending an email at the beginning of the day, with the total expense, the quantity of products purchased and the average price of the products.

Let's use pyautogui, a library for automating mouse and keyboard commands.

```python
# !pip install pyautogui
```

```python
import pyautogui
import time
# pyautogui.click -> mouse click
# pyautogui.write -> write a text
# pyautogui.press -> press a key
# pyautogui.hotkey -> press a combination of keys (ex: Ctrl + F)
# Step by Step

pyautogui.PAUSE = 1

# Step 1: Get into company's system (link)
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")

# Wait a few seconds for the browser connection to complete.
time.sleep(5)

# Setp 2: Login
# Get login field
pyautogui.click(x=973, y=426)
# Write login
pyautogui.write("meu_login")

# Same for the Pass
pyautogui.click(x=935, y=521)
pyautogui.write("minha_senha")

# Clique in Access
pyautogui.click(x=971, y=601)
time.sleep(3)

# Step 3: Export database
pyautogui.click(x=510, y=330) # Click on purchase file
pyautogui.click(x=1661, y=198) # click on the 3 dots
pyautogui.click(x=1485, y=623) # Do download
time.sleep(3)
```

```python
# # Step 4: Calculate the indicators
import pandas as pd

tabela = pd.read_csv(r"C:\Users\IGOR\Downloads\Compras.csv", sep=";")
display(tabela)
total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade
print(total_gasto)
print(quantidade)
print(preco_medio)
```

```python
import pyperclip
# Step 5: Send a mail to your boss:

# Get into mail: https://mail.google.com/mail/u/0/#inbox
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press("enter")

time.sleep(5)

# Click on Write button
pyautogui.click(x=135, y=240)

# Fill mail information
# Enter recipient
pyautogui.write("mybossmail@gmail.com")
pyautogui.press("tab") # Choose recipient

pyautogui.press("tab") # Move to subject
pyperclip.copy("Sales Report")
pyautogui.hotkey("ctrl", "v")

pyautogui.press("tab") # Move to mail body

# Using f before the text allows to insert vars
texto = f"""
Dear,
Follow the purchasing report

Total Spent: R${total_spent:,.2f}
Quantity of Products: {quantity:,}
Average Price: R${preco_medio:,.2f}

If you have any questions, do not hesitate to ask.
Att., Igor Ambrozini
"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")


# Send
pyautogui.hotkey("ctrl", "enter")
```
